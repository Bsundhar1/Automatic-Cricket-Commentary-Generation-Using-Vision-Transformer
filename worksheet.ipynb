```python
import subprocess
import time
from pyngrok import ngrok
import os

# Kill previous instances of Streamlit and ngrok (Linux/Mac)
os.system("pkill -9 streamlit || true")
os.system("pkill -f ngrok")


# Start Streamlit in the background (modify path)
process = subprocess.Popen(["streamlit", "run", "app/streamlit_app.py", "--server.port", "8501"])

# Wait for Streamlit to start
time.sleep(5)

# Start ngrok
public_url = ngrok.connect(8501, "http").public_url
print(f"🚀 Streamlit App is running at: {public_url}")

# Keep script running
try:
    while True:
        time.sleep(1)
except KeyboardInterrupt:
    print("\nShutting down...")
    process.terminate()
    ngrok.disconnect(public_url)
```
